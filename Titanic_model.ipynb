```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
df.isnull().sum()
```

```python
df=df.drop(['Cabin'], axis=1)
df.isnull().sum()
```

```python
ageIm=SimpleImputer(strategy='mean')
embarkedIm=SimpleImputer(strategy='most_frequent')
df['Age']=ageIm.fit_transform(df[['Age']])
df['Embarked']=embarkedIm.fit_transform(df[['Embarked']]).ravel()
```

```python
df.head()
```

```python
sex_map={'male':0, 'female':1}
emb_map={'S':1, 'C':2, 'Q':3}
train=df
train['Sex']=train['Sex'].map(sex_map)
train['Embarked']=train['Embarked'].map(emb_map)
```

```python
train.head(10)
```

```python
train=train.drop(['Name', 'Ticket'], axis=1)
```

```python
train.head(10)
```

```python
#train=train.drop(['Fare'], axis=1)
```

```python
predictors = train.drop(['Survived', 'PassengerId'], axis=1) 
target = train["Survived"] 
x_train, x_test, y_train, y_test = train_test_split(predictors, target, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn.metrics import accuracy_score

randomForest = RandomForestClassifier()
randomForest.fit(x_train, y_train)

y_pred = randomForest.predict(x_test)
```

```python
acc_randomforest = round(accuracy_score(y_pred, y_test) * 100, 2) 
print(acc_randomforest)
```
